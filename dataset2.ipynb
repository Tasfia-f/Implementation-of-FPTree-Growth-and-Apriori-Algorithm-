```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is loaded
# Extract the file name from the uploaded files
file_name = list(uploaded.keys())[0]

# Load CSV into Pandas DataFrame
df = pd.read_csv(file_name)

# Display the DataFrame
df.head()

# Assign a default column name (e.g., 'Item') to the first column in the DataFrame
df.columns = ['Item'] + [f'Column_{i}' for i in range(1, len(df.columns))]

# Count the frequency of each item
item_counts = df['Item'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
item_counts.plot(kind='bar', color='skyblue')
plt.title('Item Frequency Bar Plot')
plt.xlabel('Item')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Assuming 'df' is your DataFrame containing transactions
# Replace 'df' with the actual name of your DataFrame

# Convert the DataFrame to a list of lists
transactions = df.values.tolist()

# Remove 'NaN' values from the lists
transactions = [[item for item in transaction if str(item) != 'NaN'] for transaction in transactions]

# Use TransactionEncoder to one-hot encode the dataset
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

# Run Apriori algorithm
min_support = 0.1  # Adjust the support threshold as needed
frequent_itemsets = apriori(df_encoded, min_support=min_support, use_colnames=True)

# Display frequent itemsets
print("Frequent Itemsets:")
print(frequent_itemsets)

if frequent_itemsets.empty:
    print("No frequent itemsets found. Try reducing min_support.")

# Generate association rules with min_confidence set to 0.5
min_confidence = 0.5
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)

# Display association rules
print("\nAssociation Rules:")
print(rules)
```
